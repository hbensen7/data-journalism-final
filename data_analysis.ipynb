```python
## Packages
import pandas as pd
import numpy as np
```

```python
## Import data stored in Git Hub
df_raw = pd.read_csv("election-results-2012-2024_with-all-party-totals.csv")
df_raw.head(10)
```

```python
## Collapse data by FIPS code to get county-level counts
df_c = df_raw.groupby(['county_fips', 'county'], as_index = False).sum()
## Remove unnecessary columns
df_c = df_c.drop(columns = ['level', 'ctv', 'municipality', 'MCD_FIPS', 'reporting_unit', 'rep_unit_2024'])
## Round to nearest whole number
df_c = df_c.round(0)
df_c.head(10)
```

```python
# Calculate percentage democratic and republican in 2024 and 2020

df_c['PREDEMPCT24'] = df_c['PREDEM24']/df_c['USSTOT24']*100
df_c['PREREPPCT24'] = df_c['PREREP24']/df_c['USSTOT24']*100
df_c['PREDEMPCT20'] = df_c['PREDEM20']/df_c['PRETOT20']*100
df_c['PREREPPCT20'] = df_c['PREREP20']/df_c['PRETOT20']*100

# Calculate the percentage point change for dems and 
df_c['DEM_PP_CHANGE'] = df_c['PREDEMPCT24'] - df_c['PREDEMPCT20'] 
df_c['REP_PP_CHANGE'] = df_c['PREREPPCT24'] - df_c['PREREPPCT20'] 
df_c['PP_CHANGE'] = df_c['DEM_PP_CHANGE'] - df_c['REP_PP_CHANGE']
df_c['Swing_party'] = np.where(df_c['DEM_PP_CHANGE'] >= df_c['REP_PP_CHANGE'], "Democrat", "Republican")

df_c
```

```python
## Calculate new colums 

#Votes for Kamala Harris 2024 - Tammy Baldwin2024
df_c['DIFFDEM24'] = df_c['PREDEM24'] - df_c['USSDEM24']
#Votes for Donald Trump 2024 - Eric Hovde 2024
df_c['DIFFREP24'] = df_c['PREREP24'] - df_c['USSREP24'] 
#Votes for Kamala Harris 2024 - Joe Biden 2020
df_c['DIFF_PRE_DEM_24to20'] = df_c['PREDEM24'] - df_c['PREDEM20']
#Votes for Donald Trump 2024 - Donald Trump Trump 2020
df_c['DIFF_PRE_REP_20to24'] = df_c['PREREP24'] - df_c['PREREP20']



#Select the columns of interest, sort by preferred variable DIFF_PRE_REP_20to24
df_f = df_c[['county_fips', 'county', 'DIFFDEM24', 'DIFFREP24', 'DIFF_PRE_DEM_24to20', 'DIFF_PRE_REP_20to24']].sort_values(by = ['DIFF_PRE_REP_20to24'], ascending = False)

#df_f.info()
#df_f.head(10)
```

```python
# pip install censusdis

import censusdis.data as ced
# American Community Survey 5-Year Data
# https://www.census.gov/data/developers/data-sets/acs-5year.html
DATASET = "acs/acs5"

# The year we want data for.
YEAR = 2020

# This are the census variables for total population and median household income.
# For more details, see
#
#     https://api.census.gov/data/2020/acs/acs5/variables.html,
#     https://api.census.gov/data/2020/acs/acs5/variables/B01003_001E.html, and
#     https://api.census.gov/data/2020/acs/acs5/variables/B19013_001E.html.
#
TOTAL_POPULATION_VARIABLE = "B01003_001E"
MEDIAN_HOUSEHOLD_INCOME_VARIABLE = "B19013_001E"
WHITE_ALONE = "B02001_002E"

# The variables we are going to query.
VARIABLES = ["NAME", TOTAL_POPULATION_VARIABLE, MEDIAN_HOUSEHOLD_INCOME_VARIABLE, WHITE_ALONE]
```

```python
from censusdis import states

df_counties = ced.download(
    DATASET,
    YEAR,
    VARIABLES,
    state=states.WI,
    county="*",
)
df_counties.head(10)



df_counties["county_fips"] = df_counties["STATE"] + df_counties["COUNTY"]
df_counties["county_fips"] = pd.to_numeric(df_counties["county_fips"])
df_counties.info()
```

```python
df_merge = df_f.merge(df_counties, how='left')
#type(df_counties)

#Calculate the percentage of people in each county who swung toward Trump in 2024 compared to 2020
df_merge["swing_pc"] = df_merge["DIFF_PRE_REP_20to24"]/df_merge["B01003_001E"]*100

#Calcuate the percent white
df_merge["white_pct"] = df_merge["B02001_002E"]/df_merge["B01003_001E"]*100

```

```python

df_merge.to_csv('data.csv', index=False)
```
