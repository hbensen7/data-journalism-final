```python
## Packages
import pandas as pd
import numpy as np
```

```python
## Import data stored in Git Hub
df_raw = pd.read_csv("election-results-2012-2024_with-all-party-totals.csv")
df_raw.head(10)
```

```python
## Collapse data by FIPS code to get county-level counts
df_c = df_raw.groupby(['county_fips', 'county'], as_index = False).sum()
## Remove unnecessary columns
df_c = df_c.drop(columns = ['level', 'ctv', 'municipality', 'MCD_FIPS', 'reporting_unit', 'rep_unit_2024'])
## Round to nearest whole number
df_c = df_c.round(0)
df_c.head(10)
```

```python
# Calculate percentage democratic and republican in 2024 and 2020

df_c['PREDEMPCT24'] = df_c['PREDEM24']/df_c['USSTOT24']*100
df_c['PREREPPCT24'] = df_c['PREREP24']/df_c['USSTOT24']*100
df_c['PREDEMPCT20'] = df_c['PREDEM20']/df_c['PRETOT20']*100
df_c['PREREPPCT20'] = df_c['PREREP20']/df_c['PRETOT20']*100

# Calculate the percentage point change for dems and 
df_c['DEM_PP_CHANGE'] = df_c['PREDEMPCT24'] - df_c['PREDEMPCT20'] 
df_c['REP_PP_CHANGE'] = df_c['PREREPPCT24'] - df_c['PREREPPCT20'] 
df_c['PP_CHANGE'] = df_c['REP_PP_CHANGE'] - df_c['DEM_PP_CHANGE']
df_c["Republican swing"] = df_c['PP_CHANGE']/2
df_c["Democrat swing"] = df_c['PP_CHANGE']/-2
df_c['Swing_party'] = np.where(df_c['DEM_PP_CHANGE'] >= df_c['REP_PP_CHANGE'], "Democrat", "Republican")



df_c.to_csv('data_arrows.csv', index=False)

df_c.head()

```

```python
# pip install censusdis

import censusdis.data as ced
# American Community Survey 5-Year Data
# https://www.census.gov/data/developers/data-sets/acs-5year.html
DATASET = "acs/acs5"

# The year we want data for.
YEAR = 2020

# This are the census variables for total population and median household income.
# For more details, see
#
#     https://api.census.gov/data/2020/acs/acs5/variables.html,
#     https://api.census.gov/data/2020/acs/acs5/variables/B01003_001E.html, and
#     https://api.census.gov/data/2020/acs/acs5/variables/B19013_001E.html.
#
TOT_POP = "B01003_001E"
MED_HH_INC = "B19013_001E"
WHITE_ALONE = "B02001_002E"
#LAND_AREA = "ALAND"

# The variables we are going to query.
VARIABLES = ["NAME", TOT_POP, MED_HH_INC, WHITE_ALONE]

```
